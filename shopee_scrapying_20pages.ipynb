```python
from bs4 import BeautifulSoup
from selenium import webdriver 
from selenium.webdriver.chrome.options import Options
import re
import pandas as pd
import time
```

```python
options = Options()
options.add_argument("--disable-notifications")
page=1
all_data=[]
name_data=[]
offers_data=[]
brand_data=[]
price_data=[]
while page<=20:#建立迴圈抓下20頁的內容
    url='https://shopee.tw/3C%E8%88%87%E7%AD%86%E9%9B%BB-cat.69?page={}'.format((page-1))
    driver=webdriver.Chrome('/Users/liuzhengying/anaconda3/lib/python3.6/site-packages/selenium/webdriver/chrome/chromedriver')
    driver.get(url)
    soup = BeautifulSoup(driver.page_source, "html.parser")
    data_soup=soup.find_all(attrs={"type": "application/ld+json"})
    for i in data_soup:#以下為處理商品名稱,商品價格,商品品牌的list
        all_data.append(i.string)
    for i in [eval(i) for i in all_data]:
        if i.get('offers')!=None:
            offers_data.append((i.get('offers')))
        if i.get('name')!='蝦皮購物' :
            if i.get('name')!=None:
                name_data.append(i.get('name'))
        if i.get('brand')!=None:
            brand_data.append(i.get('brand'))
    for i in offers_data:
        if i.get('price')==None:
            price_data.append((i.get('lowPrice')+'~'+i.get('highPrice')))
        else:
            price_data.append(i.get('price'))
    page=page+1
    time.sleep(1)
    driver.quit()
```

```python
df=pd.DataFrame([list(a) for a in zip(name_data,brand_data,price_data)])#利用所有商品資訊創建出pandas DataFrame
df.columns=['商品名稱','品牌','價格']
df
```
